```python
#check for null values
null_values = original_dataset.isnull().sum()
```

```python
print(null_values)
```

```python
# replace null values with mean of respective columns
```

```python
# calculate means of respective columns with null values 
mean_fa = original_dataset['fixed acidity'].mean()
mean_va = original_dataset['volatile acidity'].mean()
mean_cl = original_dataset['chlorides'].mean()
mean_fsd = original_dataset['free sulfur dioxide'].mean()
mean_dty = original_dataset['density'].mean()
mean_qty = original_dataset['quality'].mean()
```

```python
# display mean values to correspond with d.p of columns

print(f'The mean for fixed acidity: {round(mean_fa,1)}')
print(f'The mean for volatile acidity: {round(mean_va, 3)}')
print(f'The mean for chlorides: {round(mean_cl, 3)}')
print(f'The mean for free sulfur dioxide: {round(mean_fsd,1)}')
print(f'The mean for density: {round(mean_dty,5)}')
print(f'The mean for Quality: {round(mean_qty,1)}')
```

```python
# substitute null values with mean 
original_dataset['fixed acidity'] = original_dataset['fixed acidity'].fillna(mean_fa)
original_dataset['volatile acidity'] = original_dataset['volatile acidity'].fillna(mean_va)
original_dataset['chlorides'] = original_dataset['chlorides'].fillna(mean_cl)
original_dataset['free sulfur dioxide'] = original_dataset['free sulfur dioxide'].fillna(mean_fsd)
original_dataset['density'] = original_dataset['density'].fillna(mean_dty)
original_dataset['quality'] = original_dataset['quality'].fillna(mean_qty)
```

```python
original_dataset_nullvalues = original_dataset.isnull().sum() 
```

```python
#confirm null values have been replaced 
print(original_dataset_nullvalues)
```

```python
#pie chart with mean values to find average amount of specific components in red wine

citric_acid = data_statistics['citric acid'].loc['mean']
residual_sugar = data_statistics['residual sugar'].loc['mean']
chlorides = data_statistics['chlorides'].loc['mean']
free_sulfur_dioxide = data_statistics['free sulfur dioxide'].loc['mean']
alcohol = data_statistics['alcohol'].loc['mean']
sulphates = data_statistics['sulphates'].loc['mean']

labels = ['Citric acid', 'Residual acid', 'Chlorides', 'Free Sulfur Dioxide', 'Alcohol', 'Sulphates']
explode_values = [0.2,0,0.3,0,0,0]

plt.pie([citric_acid, residual_sugar, chlorides, free_sulfur_dioxide, alcohol, sulphates], labels=labels, autopct= '%.2f %%', explode=explode_values)
plt.title('Components of Red Wine', fontdict= {'fontweight': 'bold'})

plt.show()
```

### The pie chart above shows the average percentage of the components of red wine using the mean values of the components.
##### The components are chlorides, residual acid, citric acid, sulphates, alcohol, and free sulfur dioxide.
